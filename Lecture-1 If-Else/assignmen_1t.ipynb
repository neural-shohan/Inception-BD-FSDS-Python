# Assignment 1: Temperature Check

```python
current_temperature = int(input("Enter current temperature: "))

if current_temperature < 10 :
    print("It's too cold!")
elif current_temperature >= 10 and current_temperature <= 25:
    print("It's cool outside.")
else:
    print("It's hot outside!")

print(current_temperature)
```

# Assignment 2: ATM Withdrawal System

```python
user_balance = int(input("Enter your balance: "))

withdraw_amount = int(input("Enter withdraw amount: "))

if withdraw_amount > user_balance:
    print("Insufficient Balance")

elif withdraw_amount  <= user_balance:
    print("Transaction successful, remaining balance:", user_balance - withdraw_amount)
```

# Assignment 3: Student Grade Calculator


```python
student_marks = int(input("Enter your marks: "))

if student_marks>= 90  and student_marks <= 100:
    print("A+")
elif student_marks >= 80 and student_marks <= 89:
    print("A")
elif student_marks >= 70 and student_marks <= 79:
    print("B")
elif student_marks >= 60 and student_marks <= 69:
    print("C")
else:
    print("Fail")
```

# Assignment 4: Bus Ticket Discount

```python
user_age = int(input("Enter your age: "))

if user_age < 5:
    print("Free")
elif user_age >= 5 and user_age <= 18:
    print("50% Discount")
elif user_age >= 60:
    print("30% Discount")
else:
    print("Full Price")
```

# Assignment 5: Restaurant Bill with Discount

```python
total_bill = int(input("Enter bill amount: "))

if total_bill > 1000:
    discount = total_bill * 0.10
    payable_amount = total_bill - discount

else:
    payable_amount = total_bill

print("Total bill is:", payable_amount)
```

# Assignment 7: Traffic Light System

```python
color = input("Enter your color: ")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

# Assignment 8: Leap Year Checker

```python
year = int(input("Enter the year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# Assignment 10: Electricity Bill Calculator

```python
units_consumed = int(input("Enter number of units: "))

if units_consumed <= 100:
    bill_amount = units_consumed * 5
elif units_consumed >= 101 and units_consumed <= 200:
    bill_amount = units_consumed * 7
else:
    bill_amount = units_consumed * 100

print("Total bill is:", bill_amount)
```
