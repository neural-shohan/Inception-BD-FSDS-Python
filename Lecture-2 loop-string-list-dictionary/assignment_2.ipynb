1. Email Validator (String Filtering + Loops)

```python
user_email = str(input("Enter email address: ")).lower().split()

for i in user_email:
    if "@" in i and i.endswith(".com") or i.endswith(".org"):
        print("Valid Email")
    else:
        print("Invalid Email")

```

2. Count Word Frequency in Customer Reviews

```python

customer_review = str(input("Enter your review: ")).lower().split()

word_count = {}

for word in customer_review:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)

```

3. Password Strength Checker

```python
user_password = input("Enter password here: ")

has_uppercase = False
has_lowercase = False
has_digit = False
has_special_character = False


for i in user_password:
    if i.isupper():
        has_uppercase = True
    elif i.islower():
        has_lowercase = True
    elif i.isdigit():
        has_digit = True
    else:
        has_special_character = True

if has_uppercase and has_lowercase and has_digit and has_special_character:
    print("Strong Password")
else:
    print("Weak Password")
```

4. Username Generator

```python
import random

user_name = str(input("Enter full name here: ")).strip().lower()

name_parts = user_name.split()

first_name = name_parts[0]

last_name = name_parts[1]


first_part = ""
count = 0 

for chacter in first_name:
    if count < 3:
        first_part += chacter
        count += 1

last_part = ""

count = 0 

index = len(last_name) - 3

while count < 3:
    last_part += last_name[index]
    index += 1
    count += 1

number = ""

for i in range(2):
    number += str(random.randint(0, 9)) 

username = first_part + last_part + number

print("Generated username:", username)




```

5. SMS Word Counter

```python
user_message = str(input("Enter message here: ")).capitalize()


word_count = len(user_message.split())
character_count = len(user_message)

if character_count > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

print(f"Words = {word_count}, Characters = {character_count}, Status  = {status}")
```

6. Remove Duplicates from Contact List

7. Detect Suspicious Words in Text (Moderation)

```python
banned_words = ['spam', 'scam', 'fake']

user_text = str(input("Enter your message: ")).lower()

warning = False

for word in banned_words:
    if word in user_text:
        warning = True
        break

if warning:
    print("Warning: Suspicious content detected")
else:
    print("Message is clean")


```

9. Count Hashtags and Mentions (Social Media Analytics)

```python
post = input("Enter your post: ")

hashtag_count = 0
mention_count = 0

words = post.split()

for word in words:
    if word.startswith("#"):
        hashtag_count += 1
    elif word.startswith("@"):
        mention_count += 1

if hashtag_count > 0 or mention_count > 0:
    print(f"Hastags = {hashtag_count}, Mentions = {mention_count}")
else:
    print("Post contains no hashtags or mentions")


```

10. Word Reversal for Data Encryption

```python
input_string = str(input("Enter your text: ")).capitalize()

words = input_string.split()

reversed_words = []

reversed_sentence = " "

for i in words:
    reversed_words.append(i[::-1])

reversed_sentence = " ".join(reversed_words)

print(reversed_sentence)




```
