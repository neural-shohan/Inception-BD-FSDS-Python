1. Inventory Price Calculator

```python
def calculate_inventory_summary(prices):
    total_cost = sum(prices)
    average_cost = total_cost / len(prices)
    return total_cost, average_cost


print(calculate_inventory_summary([120, 250, 399, 150]))
```

2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitors = set(visitor_list)
    return len(unique_visitors)

print(count_unique_visitors([101, 205, 101, 310, 205, 550]))
```

3. Product Category Counter

```python
def count_categoris(products):
    product_count = {}

    for item in products:
        if item in product_count:
            product_count[item] += 1
        else:
            product_count[item] = 1

    return product_count

print(count_categoris(["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]))
```

4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list = []
    normal_list = []

    for i in temp_list:
        if i > 38:
            high_list.append(i)
        else:
            normal_list.append(i)
    return high_list, normal_list


temperatures = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

result = analyze_temperatures(temperatures)


print(f"High List (>38 C):", result[0])
print(f"Normal List (<38 C):", result[1])
print(f"Full Output Tuple:", result)
```

5. Menu Price Lookup

```python
def get_price(menu, item):
    if item in menu:
        return menu[item]
    else:
        return "Item not found"

menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

print(get_price(menu, "Muffin"))

print(get_price(menu, "Burger"))
```

6. Student Grading System

```python
def calculate_grade(score):
    if score  >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    grades_dict = {}

    for i, j in student_dict.items():
        grades_dict[i] = calculate_grade(j)
    return grades_dict

scores = {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
grades = grade_students(scores)


print(grades)
```

7. Discount Calculator

```python
def apply_discounts(products):
    updated_price = []

    for name, price, discount in products:
        discounted_price = price * (1 - discount/100)
        updated_price.append(round(discounted_price, 2))

    return updated_price

products = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
price_after_discount = apply_discounts(products)
print(price_after_discount)


```

8. Hashtag Generator

```python
def generate_hashtags(sentence):
    words = sentence.split()
    hashtag_words  = []

    for word in words:
        hashtag_words.append("#" + word.lower())

    return hashtag_words

sentence = "AI Data Science"
hashtags = generate_hashtags(sentence)
print(hashtags)
```

10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, price):

    total = 0

    for i in cart:
        if i in price:
            total += price[i]
        else:
            print(f"{i} not found in price list")

    return total

cart = ["Milk", "Bread", "Milk", "Cheese", "Bread"]
Price = {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

total_price = calculate_total(cart, Price)

print("Total payable amount:", total_price)
```
